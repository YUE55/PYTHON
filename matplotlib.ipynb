```python
fi,axe = plt.subplots(figsize =(8,5))
x =[1,2,3,4,5,6]
y =[1,3,5,7,9,11]
np.random.seed(66)
y1 = y+np.random.randint(1,5,6)#low =0,high=5,size=5,np.random.randint(1,5,6)-取值从1-5之间的6个整数
y2 = y+np.random.randint(1,5,6)
y3 = y+np.random.randint(1,5,6)
y4 = y+np.random.randint(1,5,6)
y5 = y+np.random.randint(1,5,6)
y6 = y+np.random.randint(1,5,6)
y7 = y+np.random.randint(1,5,6)
labels = ["J","F","M","A","M","J"]
axe.stackplot(x,y1,y2,y3,y4,y5,y6,y7,labels=["1","2","3","4","5","6","7"])
axe.set_xticks(x)
axe.set_xticklabels(labels)
axe.set_title("Sales from Jan to Jun")
axe.legend(loc="upper left")
plt.show()
```

```python
#调整基本线的方法
fi,axe = plt.subplots(nrows=2, figsize =(8,5))
x =[1,2,3,4,5,6]
y =[1,3,5,7,9,11]
np.random.seed(66)
y1 = y+np.random.randint(1,5,6)#low =0,high=5,size=5,np.random.randint(1,5,6)-取值从1-5之间的6个整数
y2 = y+np.random.randint(1,8,6)
y3 = y+np.random.randint(1,5,6)
y4 = y+np.random.randint(1,30,6)
y5 = y+np.random.randint(1,5,6)
y6 = y+np.random.randint(1,20,6)
y7 = y+np.random.randint(1,10,6)
labels = ["J","F","M","A","M","J"]
axe[0].stackplot(x,y1,y2,y3,y4,y5,y6,y7, baseline="sym") #基本线将会水平对称0（the baseline is going to）
axe[0].set_xticks(x)
axe[0].set_xticklabels(labels)
axe[0].set_title("Symmetric")

axe[1].stackplot(x,y1,y2,y3,y4,y5,y6,y7, baseline="wiggle") #wiggle- 摆动的意思，重设基本线，以至于最小化平方斜率的总和（miniminzes the sum of the squared slopes）
axe[1].set_xticks(x)
axe[1].set_xticklabels(labels)
axe[1].set_title("wiggle")


plt.tight_layout()
plt.show()
```
